# Переименовали столбцы

```python
df.columns=['name','platform','year_of_release','genre','na_sales','eu_sales','jp_sales','other_sales','critic_score','user_score','rating']
df
```

# заполнили пропуски

```python
m = df['year_of_release'].median()
df['year_of_release']=df['year_of_release'].fillna(m)
```

```python
df['year_of_release']=df['year_of_release'].astype(int)
```

```python
m1 = df['critic_score'].median()
df['critic_score'] = df['critic_score'].fillna(m1)
df.isnull().sum()
```

# определили кол-во значений tbd, что расшифровывается как 'будет определено'. Можно заменить это значение на NaN и потом заменить на медианное, но мы этого делать не будем.

```python
len(df[df['user_score'] == 'tbd'])
```

```python
df[df['name'].isnull()]
```

# удалили строки 

```python
df = df.drop(659)
df = df.drop(14244)
df=df.reset_index(drop=True)
df.info()
```

# Создали новый столбец суммы продаж всех регионов

```python
df['sum_sales'] = df['na_sales']+df['eu_sales']+df['jp_sales']+df['other_sales']
```

```python
df['year_of_release'].value_counts()
```

```python
df['platform'].value_counts()
```

```python
# Построили графики согласно шагу 3
```

```python
X = df['platform']
Y = df['sum_sales']
fig, ax = plt.subplots(figsize=(16, 9))

plt.bar(X,Y,  color="purple",width=.5)
plt.title('Продажи по платформам')
plt.xlabel('Платформа')
plt.ylabel('Продажи')
plt.grid()
plt.show()
```

```python
total_platform = df.pivot_table(index='platform', values='sum_sales', aggfunc='sum').sort_values('sum_sales',ascending=False)
total_platform=total_platform.reset_index()
total_platform=total_platform.head(10)
total_platform
```

```python
X = df['platform']
Z = df['year_of_release']
fig, ax = plt.subplots(figsize=(16, 9))

plt.bar(X,Z,  color="green",width=.5)
plt.title('Продажи по платформам')
plt.xlabel('Платформа')
plt.ylabel('Продажи')
plt.ylim(1980,2020)
plt.grid()
plt.show()
```
